```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})

data.head()
```

Вариант №1

```python
lst = []
for val in data.whoAmI:
    if val == 'robot':
      lst.append([1, 0])
    else:
      lst.append([0, 1])

df = pd.DataFrame(lst, columns = ['robot', 'human'])
df.head()
```

Вариант №2

```python
for category in set(data.whoAmI):
    data[category] = [1 if category == val else 0 for val in data.whoAmI]
  
data.head()
```

Вариант №3

```python
unique_value = data.whoAmI.unique()
data[unique_value] = 0
for val in unique_value:
    data.loc[data.whoAmI == val, val] = 1
data.head()
```
