1. Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)


```python
size = ['high', 'middle', 'low']
```

```python
lst = []
for i in penguins['bill_length_mm']:
  if i > 42:
    lst.append('high')
  elif i < 35:
    lst.append('low')
  else:
    lst.append('middle')
```

```python
penguins['bill_length'] = lst
```

Как написать такой же код в одну строку

```python
penguins = sns.load_dataset('penguins').dropna() # Удаляем все na из датасета
penguins # Выводим датасет
```

```python
penguins['bill_length'] = ['high' if length >= 42 else 'middle' if length >= 35 else 'low' for length in penguins.bill_length_mm]
```
